1. Se procede con la carga de las librerías de trabajo

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

2. Se procede con la carga de datos de trabajo

```python
nxl= "/content/6. SolicitantesCrédito(USD) (1).xlsx"
XDB= pd.read_excel(nxl, sheet_name=1)
XDB= XDB[['Edad', 'Hijos', 'Perscargo', 'Estrato', 'Ingresos', 'Egresos', 'Monto (EAD)', 'Plazo', 'Cuota (COP)', 'Score' ]]
display (XDB)

XD=np.array(XDB[['Edad', 'Hijos', 'Perscargo', 'Estrato', 'Ingresos', 'Egresos', 'Monto (EAD)', 'Plazo','Cuota (COP)']])
yd=np.array(XDB[['Score']])

XDn=XD/np.max(XD,axis=0)
ydn=yd/np.max(yd,axis=0)

print("Los datos de entrada son: \n", XDn)
print("Los datos de salida son: \n", ydn)
```

3. Se crea el modelo MADALINE (Multi Adaptive Linear-Relu)

```python
madaline=tf.keras.Sequential([
    tf.keras.layers.Dense(units=10,input_shape=(9,),use_bias=False, activation='relu'),
    tf.keras.layers.Dense(units=1,use_bias=False,activation='relu')])

madaline.summary()

# Cargar pesos previamente entrenados para el modelo de Score
madaline.load_weights("/content/pesos_madaline_M.weights.h5")

madaline.compile(optimizer='sgd',loss='mse')
history2=madaline.fit(XDn,ydn,epochs=100)

#Evaluamos el modelo
yp=madaline.predict(XDn)
dfcorr2=pd.DataFrame(np.column_stack((yp,ydn)))
print("La correlación entre yp-ydn es:\n",dfcorr2.corr())

#Caida del error en el aprendizaje
ek=history2.history['loss']      #Muestra el error en el aprendizaje

plt.figure()
plt.plot(ek,'r')
plt.grid(True)
plt.show()

#Efectos independientes
WC=madaline.get_weights()
W=WC[0];C=WC[1]
a=W@C #Estos son los efectos independientes
print("Los efectos independientes son:\n",a)

for i, val in enumerate(a):
    signo = 'positivo' if val >= 0 else 'negativo'
    print(f"La variable de entrada {i+1} tiene un efecto {signo} sobre la salida.")
```

4. Evaluamos Solicitantes de Credito

```python
Xnp = [[45, 1, 0, 4, 2000, 1000, 1200, 24, 60]]
Xnpn=Xnp/np.max(XD,axis=0) #Normalizar los datos de entrada (entre 0 y 1)
print("Los datos de entrada son:\n", Xnpn)

ypnp=madaline.predict(Xnpn)
print("El score estimado es:\n", ypnp*np.max(yd))
```

5. Se crea el modelo MADALINE (Multi Adaptive Linear-Sigmoid)

```python
nxl= "/content/6. SolicitantesCrédito(USD) (1).xlsx"
XDB= pd.read_excel(nxl, sheet_name=1)
XDB= XDB[['Edad', 'Hijos', 'Perscargo', 'Estrato', 'Ingresos', 'Egresos', 'Monto (EAD)', 'Plazo', 'Cuota (COP)', 'Prob.Default (PD)' ]]
display (XDB)

XD=np.array(XDB[['Edad', 'Hijos', 'Perscargo', 'Estrato', 'Ingresos', 'Egresos', 'Monto (EAD)', 'Plazo','Cuota (COP)']])
yd=np.array(XDB[['Prob.Default (PD)']])

#Se procede con la normalización de las variables
XDn=XD/np.max(XD,axis=0)
#ydn=yd/np.max(yd,axis=0) NO ES NECESARIO NORMALIZAR LA DE SALIDA
print("Los datos de entrada son: \n", XDn)
```

```python
madaline=tf.keras.Sequential([
    tf.keras.layers.Dense(units=10, input_shape=(9,), use_bias=False, activation='relu'),
    tf.keras.layers.Dense(units=1, use_bias=False, activation='sigmoid')])

madaline.summary()

# Guardar los pesos entrenados para el modelo de Probabilidad de Default
madaline.save_weights("/content/pesos_sigmoid_M.weights.h5")

madaline.compile(optimizer='sgd', loss='mse')

history2=madaline.fit(XDn, ydn, epochs=100)

#Evaluamos el modelo
yp2=madaline.predict(XDn)
dfcorr2=pd.DataFrame(np.column_stack((yp2,ydn)))
print("La correlación entre ydn y yp es:\n",dfcorr2.corr())

#Caída del error en el aprendizaje
ek=history2.history['loss']
plt.figure()
plt.plot(ek,'r')
plt.grid(True)
plt.show()

#Efectos independientes
WC= madaline.get_weights()
W=WC[0]; C=WC[1]
a=W@C #Estos son los efectos independientes
print("Los efectos independientes son :\n",a)

for i, val in enumerate(a):
    signo = 'positivo' if val >= 0 else 'negativo'
    print(f"La variable de entrada {i+1} tiene un efecto {signo} sobre la salida.")
```

6. Evaluamos el Solicitante

```python
Xnp = [[45, 1, 0, 4, 2000, 1000, 1200, 24, 60]]
Xnpn=Xnp/np.max(XD,axis=0) #Normalizar los datos de entrada (entre 0 y 1)
print("Los datos de entrada son:\n", Xnpn)

ypnp=madaline.predict(Xnpn)
print("la Probabilidad de Default del crédito es:\n", ypnp*np.max(yd))
```

**ANALISIS DE RESULTADOS**

**Análisis del modelo MADALINE aplicado a la predicción crediticia**

El análisis se desarrolló en dos modelos neuronales tipo MADALINE, cada uno con objetivos específicos: predecir el Score crediticio y estimar la Probabilidad de Default de los solicitantes, utilizando variables socioeconómicas y financieras tales como edad, ingresos, egresos, número de hijos, estrato socioeconómico, monto solicitado y cuota mensual, entre otras.

En el primer modelo, enfocado en estimar el Score crediticio, se utilizó una arquitectura con función de activación ReLU en ambas capas (una capa oculta de 10 neuronas y una de salida con una sola neurona). Este modelo presentó un rendimiento sobresaliente reflejado en una alta correlación (superior a 0.99) entre las predicciones y los valores reales normalizados, lo que evidencia una capacidad predictiva robusta y una adecuada captura de las relaciones subyacentes entre variables. La curva de entrenamiento reveló una reducción significativa del error cuadrático medio durante las primeras épocas, seguido de una estabilización progresiva en niveles mínimos, indicando un aprendizaje eficiente y sin signos aparentes de sobreajuste.

Una ventaja destacada de este modelo es la interpretabilidad lograda a través del análisis independiente de pesos, donde se identificó claramente que variables como ingresos altos y cuotas manejables tienen efectos positivos notorios sobre el Score, mientras que variables como altos egresos podrían disminuirlo. Esto resulta especialmente valioso para instituciones financieras que buscan comprender y justificar decisiones crediticias desde una perspectiva explicativa y no meramente predictiva.

El segundo modelo, orientado a predecir la Probabilidad de Default, empleó una arquitectura similar pero con una función de activación sigmoide en la capa de salida, adaptándose así al carácter probabilístico del objetivo (valores entre 0 y 1). Aunque la correlación obtenida fue moderada (cerca de 0.47), este resultado es consistente con modelos predictivos en escenarios de riesgo crediticio, donde factores externos y el comportamiento humano introducen incertidumbres adicionales. La curva de pérdida mostró una disminución constante y efectiva del error durante las 100 épocas, indicando una correcta adaptación progresiva del modelo a los datos proporcionados.

En cuanto a la evaluación específica del solicitante, los modelos sugirieron un perfil de riesgo moderado-bajo, respaldado por indicadores financieros favorables como una relación equilibrada entre deuda e ingreso y una estabilidad financiera representada por la edad y un historial crediticio considerable. Este perfil refuerza la percepción de que el solicitante tiene condiciones adecuadas para el cumplimiento de sus obligaciones, apoyando así una potencial aprobación del crédito bajo los criterios habituales del sector financiero.

En resumen, ambos modelos neuronales han demostrado utilidad en contextos de análisis crediticio, ofreciendo predicciones sólidas acompañadas de interpretabilidad suficiente para guiar decisiones financieras estratégicas.
